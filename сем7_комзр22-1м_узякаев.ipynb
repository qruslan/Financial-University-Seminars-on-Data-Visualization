```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

PATH =  "/content/drive/MyDrive/Colab Notebooks/data/"
df = pd.read_csv(PATH + "Experiment.csv", encoding="cp1251", sep=';', decimal=',')
df.head()
```

```python
err = np.array([[-0.2, -0.12, -0.93, -0.03, -0.2, -0.1, -0.4, -0.6], 
                            [0.2, 0.05, 0.4, 0.25, 0.2, 0.2, 0.2, 0.71]])
```

```python
q = np.random.randint(0, len(df), 3)
for item in q:
  data = df.loc[item][2:]
  plt.errorbar(range(1, len(data) + 1), data, yerr = err, linestyle='none', marker="o", label="Эксперимент" + str(item), capsize=8, capthick=1)
plt.xlabel("Номер девайса")
plt.ylabel("Измерение")
plt.legend()
```

# 4

```python
import csv
df = pd.read_csv(PATH + "mlb_players.csv",sep=",", quoting=csv.QUOTE_NONE)[:1034] #последняя строчка пустая
df.columns = ["Name", "Team", "Position", "Height(inches)", "Weight(lbs)", "Age"]
df
```

```python
def remove_quotes(s):
  try:
    s = eval(s)
  except TypeError:
    pass
  return s
```

```python
df = df.applymap(remove_quotes)
df
```

```python
oak = df.loc[df['Team'] == "OAK"]
oak.head()
```

```python
plt.hist(oak["Age"], bins = 15)
plt.xlabel("Возраст")
plt.ylabel("Количество")
plt.grid(which='major')
plt.grid(which='minor', linestyle=':')
plt.minorticks_on()
```

```python
plt.rcParams["figure.figsize"] = (24,8)
plt.hist(oak["Position"], bins = 9)
plt.xlabel("Возраст")
plt.ylabel("Количество")
plt.grid(which='major')
plt.grid(which='minor', linestyle=':')
plt.minorticks_on()
```
