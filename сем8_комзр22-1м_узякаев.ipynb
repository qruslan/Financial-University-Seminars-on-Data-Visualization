# 2

```python
df = pd.read_csv(PATH + "city_russia.csv")
df.head()
```

```python
df.foundation_year
```

```python
plt.rcParams["figure.figsize"] = (8,4)
plt.hist(df["foundation_year"], bins=10);
plt.xlabel("Год основания")
plt.ylabel("Количество")
plt.grid()
```

```python
import scipy.stats as stats
mean, std = stats.norm.fit(df["foundation_year"].values)
pdf = stats.norm.pdf(sorted(df["foundation_year"].values), loc=mean, scale=std)
pdf
```

```python
stats.norm.fit(df["foundation_year"].values)
```

```python
plt.rcParams["figure.figsize"] = (8,4)
plt.hist(pdf, bins=10);
plt.xlabel("Плотность вероятности")
plt.ylabel("Количество")
plt.grid()
```

# 3

```python
plt.rcParams["figure.figsize"] = (5,5)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(sorted(df["foundation_year"].values), pdf, color='black')
ax.hist(df["foundation_year"], bins=8, density=True, color='yellow');
ax.plot([], [], ' ', label=f"Среднее:{mean}")
ax.plot([], [], ' ', label=f"Ст.откл.:{std}")
ax.plot([], [], ' ', label=f"Количество:{len(pdf)}")
ax.set_xlabel("Год основания");
ax.set_ylabel("Вероятность");
ax.set_xlim(-500, 2500)
ax.legend()
plt.grid()
```
