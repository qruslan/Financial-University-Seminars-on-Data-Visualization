```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# login: 221788
n = 8

x, y = np.mgrid[-2:3, 0:4]

def f(x,y):
  return (1 - np.sin(np.sin(y)) + x ** n) * np.exp(x ** n + y)

fig, ax = plt.subplots()

cs = ax.contourf(f(x,y), levels = n)
plt.colorbar(cs)

fig.set_figwidth(n+3)    #  ширина и
fig.set_figheight(n)    #  высота "Figure"


plt.show()
```

```python
fig, ax = plt.subplots(nrows= 2 , ncols= 2, figsize=(9,9))
x1 = np.arange(-5, 5)
y1 = np.sin(np.sin(x1)) * np.exp(x1)

x2 = np.arange(-7, 10)
y2 = -4 * n * 2 / x2 + x2 ** 3

x3 = np.arange(-25, 25)
y3 = np.sin(np.sin(x3)) * np.exp(x3)

x4 = np.arange(-35, 50)
y4 = -4 * n * 2 / x4 + x4 ** 3

ax[0, 0].plot(x1, y1)
ax[0, 1].plot(x3, y3)
ax[1, 0].plot(x2, y2)
ax[1, 1].plot(x4, y4)
```
