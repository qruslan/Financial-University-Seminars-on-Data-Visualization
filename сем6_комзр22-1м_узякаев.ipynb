```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import csv

PATH =  "/content/drive/MyDrive/Colab Notebooks/data/"
fem = pd.read_csv(PATH + "oscar_age_female.csv",sep=",", quoting=csv.QUOTE_NONE)
male = pd.read_csv(PATH + "oscar_age_male.csv",sep=",", quoting=csv.QUOTE_NONE)
```

```python
fem
```

```python
male
```

```python
fem.columns = ['Year', 'Age', 'Name', 'Movie', 'del']
male.columns = ['Year', 'Age', 'Name', 'Movie', 'del']
male = male.drop(columns=['del'], axis=1)
fem = fem.drop(columns=['del'], axis=1)
```

```python
x = np.arange(1927, 2016)
y1 = fem.Age
y2 = male.Age
```

```python
lines = plt.plot(x, y1, x, y2)
plt.legend(['female', 'male'])
plt.ylabel("Возраст")
plt.xlabel("Год")
plt.grid()
```

```python
plt.hist(male.Age, bins=8, color="red")
plt.legend(['male'])
plt.xlabel("Возраст")
plt.grid()
```

```python
plt.hist(fem.Age, bins=8, color="green")
plt.legend(['female'])
plt.xlabel("Возраст")
plt.grid()
```

# 2

```python
df = pd.read_csv(PATH + "iris.csv",sep=",", quoting=csv.QUOTE_NONE)
df
```

```python
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'variety']
df
```

```python
x1 = df.sepal_length
x2 = df.petal_length
y1 = df.sepal_width
y2 = df.petal_width

fig, ax = plt.subplots(2, figsize=(6, 9))
ax[0].scatter(x1, y1) #sepal
ax[1].scatter(x2, y2) #petal
```

```python
import seaborn as sns

sns.FacetGrid(df, hue ="variety", height=6).map(plt.scatter, 'sepal_length', 'petal_length').add_legend()
```

```python
plt.hist(df.sepal_length, bins=8)
plt.xlabel("sepal_length")
plt.grid()

```

```python
plt.hist(df.sepal_width, bins=8)
plt.xlabel("sepal_width")
plt.grid()
```

```python
plt.hist(df.petal_length, bins=8)
plt.xlabel("petal_length")
plt.grid()
```

```python
plt.hist(df.petal_width, bins=8)
plt.xlabel("petal_width")
plt.grid()
```

```python
tips = pd.read_csv(PATH + "tips.csv", sep=",")
tips.columns = ["total_bill", "tip", "sex", "smoker", "day", "time", "size"]
tips
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax.scatter(tips.total_bill, tips.tip)
plt.xlabel("total_bill")
plt.ylabel("tip")
```

```python
plt.hist(tips.tip, bins=10)
plt.grid()
```

```python
tips.groupby("day").tip.unique()
```

```python
plt.hist(tips.groupby("day").tip.unique(), bins=8)
plt.legend(["Fri", "Sat", "Sun", "Thur"])
```

```python
plt.hist(tips.groupby("sex").tip.unique(), bins=8)
plt.legend(["Female", "Male"])
```
