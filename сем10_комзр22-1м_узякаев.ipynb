```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# 1

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.1)
Y = np.arange(-1, 10, 0.1)
X,Y = np.meshgrid(X,Y)

def f(x,y):
    return 0.5 * np.sin(x) + np.cos(y ** 2) + np.sin(x)

ax.plot_surface(X, Y, f(X,Y), rstride=1, cstride=1)
```
