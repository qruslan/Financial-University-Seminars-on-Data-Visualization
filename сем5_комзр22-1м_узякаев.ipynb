```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

x = np.arange(-10, 7, 0.4)
y =  x / np.cos(x)

bot = np.ma.masked_where(y <= 0, y)
top = np.ma.masked_where(y >= 0, y)

plt.figure(figsize=(8, 4))
plt.plot(x, top, color='blue')
plt.plot(x, bot, color='red')
plt.grid()
```

```python
x = np.arange(-10, 10, 0.3)
y = np.sin(x) * (5 - x)
bot = np.ma.masked_where(y <= 0, y)
top = np.ma.masked_where(y >= 0, y)
plt.figure(figsize=(8, 4))
plt.plot(x, top, color='blue')
plt.plot(x, bot, color='red')
plt.grid()
```
